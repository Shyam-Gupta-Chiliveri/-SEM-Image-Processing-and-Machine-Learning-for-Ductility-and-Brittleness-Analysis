```python
# Load the image
image = cv2.imread("C:\\Users\\shyam\\Part 1 - bending test on Motor carrios (Supplier-A).tiff")

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply a threshold to binarize the image
_, threshold = cv2.threshold(gray, 10, 255, cv2.THRESH_BINARY)

# Find contours in the thresholded image
contours, _ = cv2.findContours(threshold, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Get the contour with the largest area
largest_contour = max(contours, key=cv2.contourArea)

# Get the bounding box of the largest contour
x, y, w, h = cv2.boundingRect(largest_contour)

# Crop the image to remove the black toolbar
cropped_image = image[y:y+h, x:x+w]

# Save or display the cropped image
cv2.imwrite('cropped_image_test.tiff', cropped_image)
```
